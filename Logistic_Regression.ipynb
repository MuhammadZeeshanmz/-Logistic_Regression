```python
# pre-processing
import numpy as np
tweet_tokens_pos = twitter_samples.tokenized('positive_tweets.json')
tweet_tokens_neg = twitter_samples.tokenized('negative_tweets.json')
all_tweets = tweet_tokens_pos + tweet_tokens_neg
labels = np.concatenate((np.ones(len(positive_tweets)), np.zeros(len(negative_tweets))))
print(tweet_tokens_pos[:3])
print(tweet_tokens_neg[:3])
```

```python
# getting stop words and punctuation
import string
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print('stop_words\n', stop_words)
print(stop_words)
print('\nPunctuation\n')
print(string.punctuation)
```

```python
# removing stopword and punctuation from tweets, and converting to lowercase
print('\033[92m')
print(all_tweets[:3])
print('\033[94m')
cleaned_tweets = []
for tweet in all_tweets:
    cleaned_tweet = []
    for word in tweet:
        if (word.lower() not in stop_words and word not in string.punctuation):
            cleaned_tweet.append(word.lower())
    cleaned_tweets.append(cleaned_tweet)
print('removed stop words and punctuation:')
print(cleaned_tweets[:3])
```

```python
# handling emojis and emoticons
# happy emoticons
emoticon_happy = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
])
# sad emoticons
emoticon_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
])
print(emoticon_happy)
print(emoticon_sad)
```

```python
import numpy as np

# ----- 1. Sigmoid function -----
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# ----- 2. Cost function -----
def compute_cost(y, y_hat):
    m = y.shape[0]
    epsilon = 1e-15  # avoid log(0)
    y_hat = np.clip(y_hat, epsilon, 1 - epsilon)  # prevent numerical instability
    cost = (-1 / m) * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    return cost


# ----- 3. Gradient computation -----
def compute_gradients(X, y, y_hat):
    m = y.shape[0]
    dw = (1 / m) * np.dot(X.T, (y_hat - y))
    db = (1 / m) * np.sum(y_hat - y)
    return dw, db


# ----- 4. Gradient Descent training loop -----
def gradient_descent(X, y, weights, bias, epochs, learning_rate=0.01):
    costs = []

    for i in range(epochs):
        # Linear model
        z = np.dot(X, weights) + bias

        # Sigmoid activation
        y_hat = sigmoid(z)

        # Cost
        cost = compute_cost(y, y_hat)

        # Gradients
        dw, db = compute_gradients(X, y, y_hat)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        costs.append(cost)
        if i % 100 == 0 or i == epochs - 1:
            print(f"Epoch {i}, Cost: {cost:.4f}")

    return weights, bias, costs


# ----- 5. Training -----
# Convert sparse matrix to dense (if needed)
X_train_dense = X_train_vectorized.toarray()

# Ensure y_train is a numpy array and reshaped properly
y_train = np.array(y_train)
if y_train.ndim == 2:
    y_train = y_train.ravel()  # flatten if needed

# Initialize weights and bias
n_features = X_train_dense.shape[1]
weights = np.random.randn(n_features) * 0.01
bias = 0.0

# Hyperparameters
epochs = 1000
learning_rate = 0.01

# Start training
weights, bias, costs = gradient_descent(
    X_train_dense, y_train, weights, bias, epochs, learning_rate
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize and train the logistic regression model
model = LogisticRegression(max_iter=1000)  # You can adjust max_iter as needed
model.fit(X_train_array, y_train)  # Train the model

# Making predictions for the training and test sets
y_pred_train = model.predict(X_train_array)  # Make predictions on the training set
y_pred_test = model.predict(X_test_array)    # Make predictions on the test set

# Evaluation metrics
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print("\nTraining Classification Report:\n", classification_report(y_train, y_pred_train))
print("Test Classification Report:\n", classification_report(y_test, y_pred_test))

# Confusion Matrices
train_conf_matrix = confusion_matrix(y_train, y_pred_train)
test_conf_matrix = confusion_matrix(y_test, y_pred_test)

print("\nTraining Confusion Matrix:\n", train_conf_matrix)
print("Test Confusion Matrix:\n", test_conf_matrix)

# Plotting confusion matrix for the test set with improved styling
plt.figure(figsize=(7, 6))
sns.heatmap(test_conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Pred 0", "Pred 1"], yticklabels=["Actual 0", "Actual 1"],
            cbar_kws={'label': 'Number of Predictions'})
plt.title("Confusion Matrix - Test Set", fontsize=14)
plt.xlabel("Predicted", fontsize=12)
plt.ylabel("Actual", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
